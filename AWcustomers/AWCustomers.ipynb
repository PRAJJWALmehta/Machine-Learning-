## 1. Feature Selection

```python
df_raw = pd.read_csv('https://raw.githubusercontent.com/PRAJJWALmehta/Machine-Learning-/main/Data/AWCustomers.csv')
print(df_raw.shape)
df_raw.head()
```

```python
df_raw.describe()
```

```python
# # a) feature selection 
features = ['Education', 'Occupation', 'Gender', 'MaritalStatus', 'HomeOwnerFlag', 'NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
# features that are relevant for predcition of Bike sales

# b)
df = df_raw[features]
df.head()
```

Birthdate is an inconvenient format, therefore we will convert it to Age

```python
from datetime import date
df_raw.BirthDate = pd.to_datetime(df_raw.BirthDate)
df_raw.head()
```

```python
# converting birthdate to age

def calculate_age(birthdate):
  today = date.today()
  age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
  return age

birthdate = df_raw['BirthDate']
Age = birthdate.apply(calculate_age)
df['Age'] = Age
df.head()

```

## 2. Data Cleaning and Pre processing

### Handling Null values

In order to replace Null values we first need to determine their frequency and type

Since there are no Null values, we can proceed further

```python
df.head()
```

### Normalization

```python
df.head()
```

```python
df.std()
```

the column that needs to be standardized is the Yearly salary column 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

cols = ['YearlyIncome', 'Age']
df['YearlyIncome'] = scaler.fit_transform(df[cols])[:, 0]
df['Age'] = scaler.fit_transform(df[cols])[:, 1]
df.head()
```

### Standardization

### Discretization

Data discretization is the process of converting continuous data into discretebuckets by grouping it


```python
# since all our data is already discrete, there is no need for discretization
```

### Label Encoding

Label encoding is done in order to make ordinal or categorical data easy to process by assigning them a numerical value

```python
df['Education'].value_counts()
```

```python
edu = df['Education'].map({'Partial High School': 1, 'High School' : 2, 'Partial College' : 3, 'Bachelors' : 4, 'Graduate Degree' : 5})
df['Education'] = edu
```

```python
df['Occupation'].value_counts()
```

```python
df['Occupation'] = df['Occupation'].map({'Manual' : 1, 'Skilled Manual' : 2, 'Clerical' : 3, 'Management' : 4, 'Professional' : 5})
```

```python
df['Gender'] = df['Gender'].map({'M' : 0, 'F' : 1})
```

```python
df['MaritalStatus'] = df['MaritalStatus'].map({'M' : 0, 'S' : 1})
```

## 3. Correlation Analysis

```python
from scipy.spatial import distance
```

```python
# using cosine algorithm to compute similarity between gender and yearly income
dist1 = distance.cosine(df['Gender'].values, df['YearlyIncome'].values)
dist1
```

```python
# using jaccard similarity algorithm to compute proximity
dist2 = distance.jaccard(df['Occupation'].values, df['YearlyIncome'].values)
dist2
```
